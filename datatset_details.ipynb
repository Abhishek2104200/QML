```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Load the dataset
import pandas as pd
import numpy as np
import os

# Provide the full path to your CSV file in Drive
file_path = "/content/drive/My Drive/sequences (1).csv"

# Load the dataset
try:
    df = pd.read_csv(file_path)
    print(f"‚úÖ Successfully loaded: {file_path}")
except Exception as e:
    raise ValueError(f"‚ùå Could not load the dataset. Error: {e}")

# Step 3: Dataset Exploration

# === Basic Info ===
print("\n======== DATASET OVERVIEW ========")
print(f"üìÑ File Name: sequences (1).csv")
print(f"üî¢ Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")

# === Column Data Types ===
print("\n======== COLUMN DATA TYPES ========")
print(df.dtypes)

# === Missing Values ===
print("\n======== MISSING VALUES PER COLUMN ========")
print(df.isnull().sum())

# === Sample Data ===
print("\n======== FIRST 5 ROWS OF THE DATASET ========")
print(df.head())

# === Descriptive Stats: Numerical Columns ===
print("\n======== NUMERICAL SUMMARY ========")
print(df.describe(include=[np.number]))

# === Descriptive Stats: Categorical Columns ===
print("\n======== CATEGORICAL FEATURES SUMMARY ========")
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    print(f"\n-- Column: {col} --")
    print(df[col].value_counts().head(10))
    print(f"üîπ Unique values: {df[col].nunique()}")

# === Sequence Length Stats (Text Columns) ===
print("\n======== SEQUENCE LENGTH ANALYSIS (OBJECT COLUMNS) ========")
for col in cat_cols:
    try:
        lengths = df[col].dropna().apply(len)
        if lengths.mean() > 10:
            print(f"\nüß¨ Column: {col}")
            print(f"   ‚û§ Average Length: {lengths.mean():.2f}")
            print(f"   ‚û§ Min Length: {lengths.min()}")
            print(f"   ‚û§ Max Length: {lengths.max()}")
    except Exception as e:
        print(f"‚ö†Ô∏è Skipping column '{col}' due to error: {e}")
```
